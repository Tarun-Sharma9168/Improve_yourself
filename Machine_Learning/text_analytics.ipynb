```python
#counting the number of words in the financial report
def count_words(text):
    tokenized_text=word_tokenize(text)
    return len(tokenized_text)
```

```python
#counting the number of sentences 
def count_sentences(text):
    tokenized_sent=nltk.sent_tokenize(text)
    return len(tokenized_sent)
```

```python
#Average word length measured in the text
def average_word_length(text):
    words=text.split()
    average=sum(len(word) for word in words)/len(words)
    return average
```

```python
#finding the positive score for the document 
def positive_score(positive_word_list):
    result=len(positive_word_list)
    return result
```

```python
#finding the negative score for the document 
def negative_score(negative_word_list):
    result=len(negative_word_list)
    return result
```

```python
#range is between -1 to +1
#getting out with the polarity scores
def polarity_score(positive_score,negative_score):
    result=(positive_score-negative_score)/((positive_score+negative_score)+0.000001)
    return result
```

```python
#Range is from 0 to 1
#getting out with the subjectivity_score 
def subjectivity_score(positive_score,negative_score,filtered_word_count):
    result=(positive_score+negative_score)/((filtered_word_count)+0.000001)
    return result
```

```python
#finding the sentiment_core_categorization
def sentiment_core_categorization(polarity_score):
    if(polarity_score < -0.5):
        return 'Most Negative'
    elif (polarity_score >= -0.5 and polarity_score < 0):
        return 'Negative'
    elif (polarity_score == 0):
        return 'Neutral'
    elif (polarity_score > 0 and polarity_score <=0.5):
        return 'Positive'
    else:
        return 'Very Positive'
```

```python
#finding the average sentence length 
#it is for the single url or we can say that the single financial report that we are considering 
def avg_sentence_length(num_of_words,num_of_sentences):
    result=num_of_words/num_of_sentences
    return result
```

```python
#finding the percentage of the complex words in the document
def percentage_of_complex_words(num_complex_words,num_words):
    result=num_complex_words/num_words
    return result
```

```python
#finding the fog index 
def fog_index(avg_sentence_length,percentage_of_complex_words):
    result=(0.4)*(avg_sentence_length+percentage_of_complex_words)
    return result
```

```python
#finding the list of the words which are neutral, positive and the negative
def finding_positive(test_subset):    
    #test_subset=['20170412', 'great', 'bad', 'terrible', 'dog', 'stop', 'good']
    sid = SentimentIntensityAnalyzer()
    pos_word_list=[]
    neu_word_list=[]
    neg_word_list=[]
    

    for word in test_subset:
        if (sid.polarity_scores(word)['compound']) >= 0.5:
            pos_word_list.append(word)
        elif (sid.polarity_scores(word)['compound']) <= -0.5:
            neg_word_list.append(word)
        else:
            neu_word_list.append(word)                

    return (len(pos_word_list),len(neu_word_list),len(neg_word_list))
```

```python
#finding the words are complex or not in the text
def complex_words_or_not(test_subset):
    english_most_common_10k = 'https://raw.githubusercontent.com/first20hours/google-10000-english/master/google-10000-english-usa-no-swears.txt'

    # Get the file of 10 k most common words from TXT file in a github repo
    response = requests.get(english_most_common_10k)
    data = response.text

    set_of_common_words = {x for x in data.split('\n')}

    # Once we have the set of common words, we can just check.
    # The check is in average case O(1) operation,
    # but you can use for example some sort of search three with O(log(n)) complexity
    #while True:
    #    word = input()
    #    if word in set_of_common_words:
    #        print(f'The word "{word}" is common')
    #    else:
    #        print(f'The word "{word}" is difficult')
    
    
    complex_words=[x for x in test_subset if x not in set_of_common_words]
    return len(complex_words)
```
